```python
import pandas as pd 
from sqlalchemy import create_engine
import sqlite3
```

```python
comm = sqlite3.connect("data.db")
cursor = comm.cursor()
```

LIST THE TABLES AVAILABLE IN THE DATABASE [data.db]

```python
cursor.execute("""
    SELECT name
    FROM sqlite_master
    WHERE type = 'table';
""")
print(cursor.fetchall())
```

```python
engine = create_engine("sqlite:////content/data.db")
```

LET US QUERY THE [hpd311calls] TABLE 

```python
query1 = """
    SELECT * 
    FROM hpd311calls
"""

all_hp = pd.read_sql(query1, engine)
```

```python
all_hp.head(3)
```

```python
query2 = """
    SELECT * 
    FROM hpd311calls
    WHERE borough = 'BROOKLYN'
"""

brooklyn_bou = pd.read_sql(query2, engine)
```

```python
brooklyn_bou.head(2)
```

```python
query3 = """
    SELECT *
    FROM weather
"""
weather =  pd.read_sql(query3, engine)
```

```python
weather.head(2)
```

```python
query4 = """
    SELECT *
    FROM hpd311calls hp
    JOIN weather w
        ON hp.created_date = w.date
"""
join = pd.read_sql(query4, engine)
```
